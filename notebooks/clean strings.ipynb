```python
import pandas as pd

df = pd.read_csv("../data/messy_strings.csv")

def clean_strings(strings):
    df["raw"] = df["raw"].dropna()
    df["raw"] = df["raw"].astype(str)
    df["raw"] = df["raw"].str.strip()
    df["raw"] = df["raw"].str.lower()
    df["raw"] = df["raw"].str.replace(r'[^a-zA-Z0-9\s]', '', regex=True)
    return df["raw"]

df["clean"] = clean_strings(df["raw"])
print("# Rows:", len(df["clean"]))
print("Unique count:", df["clean"].nunique())
print("Common items:", df["clean"].value_counts())

df.to_csv('../data/messy_strings_clean.csv', index=False)
    
df["clean"].value_counts().head(5).plot.bar()
```
